### Q 2.5: Creating the Dataframe
#### Finally, use the collected random facts to create a data frame. Your header should be : ( id, text, source, source_url, language, and permalink). Display the data frame.

```python
random_facts_URL = 'https://uselessfacts.jsph.pl/api/v2/facts/random'
page = requests.get(random_facts_URL)

data = page.json()
header = data.keys()

t = 2
random_facts = []
for i in range(10):
    page = requests.get(random_facts_URL)
    data = page.json()
    values = data.values()
    random_facts.append(values)
    time.sleep(t)
    
random_facts_df = pd.DataFrame(random_facts, columns=header)
random_facts_df
```
